# 3. TICKET FARE AND CLASS

```python
# Distribution of fares paid by passengers

data['Fare'].describe()
```

```python
# Fare Distribution Chart

data['Fare'].hist(bins=5, grid = False);

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency");
```

```python
# Fares of passengers range from 0 to 512 naira with 0 to 100 naira having the highest frequency.
```

```python
# What is the distribution of Fares paid by passengers across different classes?
# Grouping by class

class_labels = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
for class_val, val in data.groupby('Pclass')['Fare']:
    plt.hist(val, alpha=0.5, bins=5, label=class_labels[class_val])

plt.title("Fare Distribution by Class")
plt.xlabel('Fare')
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# People in first class paid 0 to 512 naira
# People in second class paid 0 to 75 naira
# People in third class paid 0 to 72 naira
# The chart also shows that people in third class were more in number
```

```python
# How does the fare distribution vary among differnt classes?
data.groupby('Pclass')['Fare'].sum()
```

```python
# People in first class paid more 
```

```python
# How does the Age distribution differ between classes?

class_labels = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
for class_val, val in data.groupby('Pclass')['Age']:
    plt.hist(val, alpha=.5, label=class_labels[class_val])

plt.title("Age Distribution By Class")
plt.xlabel('Age')
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# For First class, Age range from 2 to 80 yrs
# For Second class, Age range from 2 to 70 yrs
# For Third class, Age range from 1 to 74 yrs 
```

```python
# Is there any relationship between Ticket fare and Survival?

import seaborn as sns
plt.figure(figsize = (10,8))
sns.heatmap(data[['Survived', 'Fare']].corr(), annot = True, cmap = 'viridis')

plt.title('Features Correlation Heatmap')
```

```python
# There is positive correlation (0.23) between Fare and Survival
```

# 4. FAMILY RELATIONSHIPS

```python
# What proportion of passengers traveled with family members?

proportion_with_family = (data[data['Family'] > 0].shape[0] / data.shape[0]) * 100
print(f"Proportion of Passengers with family members: {proportion_with_family:.4f}%")
```

```python
# Does having family aboard impact survival rates?
# How does the Sutvival rate differ between passengers with and without family?

survival_with_family = data[data['Family'] > 0]['Survived'].mean() * 100
survival_without_family = data[data['Family'] == 0]['Survived'].mean() * 100
print(f"survival rate for people with family:     {survival_with_family:.2f}%")
print(f"survival rate for people without family:  {survival_without_family:.2f}%")
```

```python
# People with fammily have a high survival rate of 50.67%
```
